```python
import pandas as pd
trump_tweets = pd.read_csv('realDonaldTrump.csv',encoding='mac_roman')
obama_tweets = pd.read_csv('BarackObama.csv',encoding='mac_roman')
```

```python
import nltk
from nltk.sentiment import vader
nltk.download('vader_lexicon')
nltk.download('stopwords')
```

```python
sia = vader.SentimentIntensityAnalyzer()
```

```python
#create new column and add Vader sentiment scores

trumpPolarity = []
for i in trump_tweets['text']:
    score = sia.polarity_scores(i)['compound']
    trumpPolarity.append(score)
    
trump_tweets['Polarity'] = trumpPolarity
```

```python
#Calculate Mean of dataframe
trump_tweets['Polarity'].mean()
```

```python
#Calculate Standard dev of dataframe

trump_tweets['Polarity'].std()
```

```python
#Make histogram of polarity scores

trump_tweets['Polarity'].hist()
```

```python
#Print example of most positive tweet

trump_tweets[trump_tweets['Polarity']==trump_tweets['Polarity'].max()]
```

```python
#Print example of most negative tweet

trump_tweets[trump_tweets['Polarity']==trump_tweets['Polarity'].min()]
```

```python
#create new column and add Vader sentiment scores

obamaPolarity = []
for i in obama_tweets['text']:
    score = sia.polarity_scores(i)['compound']
    obamaPolarity.append(score)
    
obama_tweets['Polarity'] = obamaPolarity
```

HOW THE TWO SETS DIFFER

Obama tweets have a majority of slightly negative tweets while Trump has mostly slightly negative and positive tweets.

Trump's most negative tweet is more negative than Obama's most negative tweet.

```python
#GENSIM AND LDA FOR TRUMP_TWEETS
from sklearn.datasets import fetch_20newsgroups
data = fetch_20newsgroups(remove=("headers", "footers", "quotes"))
```

```python
x1 = trump_tweets['text']
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
[w for w in word_tokenize(x1[0].lower()) if w not in myStopWords]
```

```python
docs = []
for i in x1:
    docs.append([w for w in word_tokenize(i.lower()) if w not in myStopWords])
```

```python
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()

docs_stemmed = []
for i in docs:
    docs_stemmed.append([p_stemmer.stem(w) for w in i])
```

```python
docs_stemmed[0]
```

```python
from gensim import corpora, models
import gensim
dictionary = corpora.Dictionary(docs_stemmed)
dictionary.filter_extremes(no_below=10, no_above=0.5)

corpus = [dictionary.doc2bow(text) for text in docs_stemmed]
for i in corpus[30]:
    print(dictionary[i[0]], i[1])
```

```python
#num_topics = 20, passes = 5
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=20, 
                                           id2word = dictionary, 
                                           passes=5)
for i in ldamodel.print_topics(num_topics=20, num_words=20):
    print(i[0])
    print(i[1])
    print('\n')
```

```python
import matplotlib.pyplot as plt
import re

re.split(re.escape(' + ') + '|' + re.escape('*'), 'hi + me*4')

fig,ax = plt.subplots(5,4,figsize=(15,20))
ax = ax.flatten()
for i in ldamodel.print_topics(num_topics=20, num_words=20):
    x = []
    y = []
    count = 0
    for j in re.split(re.escape(' + ') + '|' + re.escape('*'), i[1]):
        if count % 2 == 0:
            y.insert(0,float(j))
        else:
            x.insert(0,j)
        count += 1
    ax[i[0]].barh(x,y,height=0.5)
plt.tight_layout()
```

```python
#num_topics = 40, passes =10
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=40, 
                                           id2word = dictionary, 
                                           passes=10)
for i in ldamodel.print_topics(num_topics=40, num_words=20):
    print(i[0])
    print(i[1])
    print('\n')
```

```python
import matplotlib.pyplot as plt
import re

re.split(re.escape(' + ') + '|' + re.escape('*'), 'hi + me*4')

fig,ax = plt.subplots(10,4,figsize=(15,40))
ax = ax.flatten()
for i in ldamodel.print_topics(num_topics=40, num_words=20):
    x = []
    y = []
    count = 0
    for j in re.split(re.escape(' + ') + '|' + re.escape('*'), i[1]):
        if count % 2 == 0:
            y.insert(0,float(j))
        else:
            x.insert(0,j)
        count += 1
    ax[i[0]].barh(x,y,height=0.5)
plt.tight_layout()
```
