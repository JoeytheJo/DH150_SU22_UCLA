```python
import sklearn.datasets
import pandas as pd

diabetes = sklearn.datasets.load_diabetes()
```

```python
print(diabetes.DESCR)
```

```python
x = diabetes.data
y = diabetes.target
```

```python
#Make a scatter plot using just one of the features in x vs the target variable y
age = x[:,0] #getting the first feature's arrays
```

```python
diabetesDict = {
    'Age' : age,
    'Target' : y
} #create dictionary with age and target data

df1 = pd.DataFrame(diabetesDict) #create data frame for scatter plot
df1.plot.scatter(x = 'Age', y = 'Target') #create scatter plot
```

```python
#More plots with other features
#adding other features to dictionary
sex = x[:,1]
bmi = x[:,2]
tc = x[:,3]
ldl = x[:,4]
hdl = x[:,5]
tch = x[:,6]
ltg = x[:,7]
glu = x[:,8]

diabetesDict = {
    'Age' : age,
    'Sex' : sex,
    'BMI' : bmi,
    'TC' : tc,
    'LDL' : ldl,
    'HDL' : hdl,
    'TCH' : tch,
    'LTG' : ltg,
    'GLU' : glu,
    'Target' : y
} 
```

```python
df3 = pd.DataFrame(diabetesDict)
df3.plot.scatter(x = 'BMI', y = 'Target')
```

```python
df3.plot.scatter(x = 'TC', y = 'Target')
```

```python
df3.plot.scatter(x = 'LDL', y = 'Target')
```

```python
df3.plot.scatter(x = 'TCH', y = 'Target')
```

```python
df3.plot.scatter(x = 'LTG', y = 'Target')
```

```python
#GLU has GOOD RELATIONSHIP with y for classification
df3.plot.scatter(x = 'GLU', y = 'Target')
```

```python
#Do regression with linear regression for GLU
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model
import sklearn.model_selection

#change data into a 2-D array format
x = df3[['GLU']]
y = df3[['Target']]
```

```python
model = sklearn.linear_model.LinearRegression()
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
#x_test = x_test.reshape(-1,1)
#x_train = x_train.reshape(-1,1)
```

```python
x_test.shape
```

```python
# Train the model
model.fit(x_train,y_train)
```

```python
#Print the coefficients of model 
print(model.coef_)
```

```python
# Make a prediction
x_sample = x_test
y_pred = model.predict(x_test)
```

```python
#Scatter plot and linear fit for training data
plt.plot(x_train,y_train,'bo')

x_new = np.linspace(-0.15,0.15,100000)
x_new = x_new.reshape(-1,1)

y_pred = model.intercept_ + model.coef_ * x_new

plt.plot(x_new, y_pred)

plt.show()
```

```python
#Scatter plot and linear fit for test data

plt.plot(x_test,y_test, 'ko')

x_new = np.linspace(-0.15,0.15,100000)
x_new = x_new.reshape(-1,1)

y_pred = model.intercept_ + model.coef_ * x_new

plt.plot(x_new, y_pred)

plt.show()
```

```python
#Mean Squared Error and R-squared values for fit
from sklearn.metrics import mean_squared_error, r2_score

print('MSE = ', mean_squared_error(y_test, model.predict(x_test)))
print('R^2 = ', r2_score(y_test, model.predict(x_test)))
```
